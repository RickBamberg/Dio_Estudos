## Redução Dimensional de Imagens

### Importações

```python
import cv2
import matplotlib.pyplot as plt
```

### Carregar a imagem colorida

```python
caminho_imagem = 'lena.png' # Substitua pelo nome do seu arquivo de imagem
imagem_colorida = cv2.imread(caminho_imagem)
```

### Converter Imanges

```python
# Converter a imagem de BGR 
imagem_colorida_rgb = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

# 1. Converter para tons de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

# 2. Binarizar a imagem em tons de cinza
limiar, imagem_binaria = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)
```

### Exibir as Imagens

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagem_colorida_rgb)
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Tons de Cinza')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(imagem_binaria, cmap='gray')
plt.title('Imagem Binarizada')
plt.axis('off')

plt.show()
```
